```python
#create Agegroup
bins = [0,19,59 , df["Age"].max()]
labels = ["Teenager","Adult","Old_Age"]
df["Agegroup"] = pd.cut(df["Age"], bins = bins , labels = labels , right = True)

#cut() - splits a continous variable into discrete intervals, bins =define numerical range , labels names ech group , right inudes exremes
```

```python
#create siblings onborad group
#df["SibSp"].value_counts().sort_index() - to check the data included in it
bins = [-1,0,2,df["SibSp"].max()] #-1 to make sure 0 siblings get included
labels = ["No_siblings","ave_siblings","higher_siblings"]
df["SibSp_group"] = pd.cut(df["SibSp"], bins = bins , labels = labels , right = True)
df
```

```python
#create fare amount groupings
#df["Fare"].value_counts()#- to check the data included in it
#the fare column in skewed 
df["FareGroup"] = pd.qcut(df["Fare"] , q =3 , labels = ["Low","Medium","High"])
df
```

```python
#df.drop("Ticket" , axis = 1 ,inplace =  True) #to drop ticket
df
```

```python
#df.groupby("column_to_group")["column_to_aggregate"].aggregation_func
#e.g .mean(),.sum(),.count(),median(),.max()
```

```python
# survival  by pclass
survival_pclass = df.groupby("Pclass")["Survived"].mean() * 100
survival_pclass
#the 3rd class had the highest survival rate
```

```python
#survival by sex
survival_sex = df.groupby("Sex")["Survived"].sum()
survival_sex
```

```python
#survival by agegroup
survival_age = df.groupby("Agegroup")["Survived"].mean()  * 100
survival_age
```

```python
#to check based on number of siblings in_board
sibSp_survive = df.groupby("SibSp")["Survived"].mean()  * 100
sibSp_survive
```

```python
#to check based on fare_groupings
fare_survive = df.groupby("FareGroup")["Survived"].mean()  * 100
fare_survive
```

```python
#Multiple f=groupings
survival_age_cls = df.groupby(["Pclass","Agegroup"])["Survived"].mean()  * 100
survival_age_cls
```

```python
#based on pclass and fare
survival_cls_fare = df.groupby(["Pclass","FareGroup"])["Survived"].sum()
survival_cls_fare
```

```python
##multiple agggregations
df.groupby("Pclass").agg({
    "Survived":"mean",
    "Fare" : "mean",
    "Age" : "median"
})
```

```python
#based on location of  	Embarked
# c-cherboung(france)
# Q- Queenstown(ireland)
# s - southampton(england)
embarked_survival = df.groupby("Embarked")["Survived"].mean()  * 100
embarked_survival

pd.crosstab(df["Embarked"],df["Pclass"])
```

# Insights Based on the data

```python
1. Gender Dominance Effect
Male passengers exhibited significantly higher survival rates compared to females, likely due 
to prioritization during evacuation, indicating that gender was one 
of the most decisive survival factors rather than random chance.
2. Socioeconomic Inequality (Class Effect)
Third-class passengers had substantially higher survival rates than first classes, 
suggesting that socioeconomic status influenced access to lifeboats and evacuation priority, 
highlighting systemic inequality in survival outcomes.
3. Wealth as a Survival Proxy (Fare)
Passengers with medium to higher fares showed increased survival rates, 
reinforcing that wealth (as reflected by ticket price) was strongly correlated with survival, 
likely due to better cabin location and priority access during evacuation.  
4. Age-Based Survival Patterns
Adults passengers demonstrated higher survival rates, 
likely due to evacuation prioritization policies, 
indicating that age played a protective role in survival probability.    
5. Family Size Impact (SibSp)
Passengers with no or small family groups had higher survival rates compared to 
those with large family groups, suggesting that larger families may have faced coordination 
challenges or delays during evacuation, reducing survival chances.    
6.Embarkation as an Indirect Factor
Passengers embarking from Queenstown(ireland) exhibited higher survival rates; however, this 
appears to be driven by a higher concentration of first-class passengers from that port, 
indicating that embarkation location is not a direct cause but a proxy for socioeconomic status.    

In conclusion,
Survival aboard the Titanic was not random but strongly influenced by 
structural and social factors, particularly gender, socioeconomic status, and age. 
Male passengers, 3rd-class individuals, and wealthier to middle class passengers had significantly 
higher survival probabilities, suggesting that access to resources and prioritization during evacuation played a critical role. 
Additionally, smaller family groups had an advantage, likely due to easier mobility and coordination. Overall, the data reveals a clear 
pattern of inequality in survival outcomes, driven by both policy and social hierarchy.
    
```
